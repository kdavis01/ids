```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
bikeData = pd.read_csv('../myProject/AustinBicycleCrashes.csv')
```

```python
print(bikeData.shape)
print(bikeData.columns)
```

## Cleaning up the column names

```python
bikeData.columns = bikeData.columns.map(lambda x: x.replace(' ', '_'))
bikeData = bikeData.rename(columns={'Crash_Incapacitating_Injury_Count': 'Major_injuries', 
                                    'Crash_Possible_Injury_Count':'Possible_Injuries', 'Crash_Not_Injured_Count': 'Non_Injuries',
                                   'Crash_Non-incapacitating_Injury_Count': 'Minor_Injuries', 'Crash_Total_Injury_Count': 'Total_Injuries', 
                                   'Crash_Unknown_Injury_Count': 'Unknown_Injuries'})


print(bikeData.columns)
```

```python
bikeDataCleaned = bikeData.filter(['Crash_ID', 'Case_ID', 'Crash_Death_Count', 'Major_Injuries', 'Minor_Injuries', 'Non_Injuries',
                          'Possible_Injuries', 'Crash_Severity', 'Crash_Time', 'Total_Injuries', 'Unknown_Injuries', 'Crash_Year',
                          'Day_of_Week', 'First_Harmful_Event', 'Latitude', 'Light_Condition', 'Longitude', 'Manner_of_Collision', 
                          'Road_Class', 'Roadway_Alignment', 'Roadway_Part', 'Speed_Limit', 'Surface_Condition', 'Traffic_Control_Type', 
                          'Weather_Condition', 'Person_Helmet'])
print(bikeDataCleaned.shape)
print(bikeDataCleaned.head(30))
```
